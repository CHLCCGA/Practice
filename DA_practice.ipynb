The first three columns are year, month and day.


```python
# parse_dates gets 0, 1, 2 columns and parses them as the index
data_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data'
data = pd.read_csv(data_url, sep = "\s+", parse_dates = [[0,1,2]]) 
```

```python
data
```

```python
def fix(x):
    year = x.year - 100 if x.year > 1989 else x.year
    return datetime.date(year,x.month,x.day)

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix)

data
```

```python
data.info()
```

```python
data['Yr_Mo_Dy'] = pd.to_datetime(data['Yr_Mo_Dy'])
data = data.set_index('Yr_Mo_Dy')
```

```python
data.info()
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.notnull().sum()
```

```python
data.shape
```

```python
data.describe(percentiles=[])
```

```python
data.describe()
```

```python
day_stats = pd.DataFrame()

day_stats['min'] = data.min(axis = 1) # min
day_stats['max'] = data.max(axis = 1) # max 
day_stats['mean'] = data.mean(axis = 1) # mean
day_stats['std'] = data.std(axis = 1) # standard deviations

day_stats
```

```python
data.loc[data.index.month== 5]
```

```python
data.groupby(data.index.to_period('M')).mean()
```

```python
data.groupby(data.index.to_period('W')).mean()
```

```python
weekly =  data.resample('Y').agg(['min','max','mean','std'])
weekly.loc[weekly.index[1:10],'RPT':'ROS'].head()
```

#### EX7 - Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# set the graphs to show in the jupyter notebook
%matplotlib inline

# set seaborn graphs to a better style
sns.set(style="ticks")
```

```python
path = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Online_Retail/Online_Retail.csv'

online_rt = pd.read_csv(path, encoding = 'latin1')
```

```python
online_rt
```

```python
countries = online_rt.groupby('Country').sum()
countries = countries.sort_values(by='Quantity',ascending=False)

countries['Quantity'].plot(kind='bar')

plt.xlabel('Countries')
plt.ylabel('Quantity')
plt.title('Countries with most orders')

plt.show()
```
